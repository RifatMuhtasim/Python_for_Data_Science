```python
import pandas as pd
```

```python
## Import Data from CSV
df = pd.read_csv("../../data/survey_results_public.csv", index_col="Respondent")
schema_df = pd.read_csv("../../data/survey_results_schema.csv", index_col="Column")
```

```python
## filtering Bangladesh Data
filt = df['Country'] == "Bangladesh"
bangladesh_df = df.loc[filt]
bangladesh_df.head()
```

```python
## Export Bangladesh_df as a CSV
bangladesh_df.to_csv("data/bangladeshi_developers_survey.csv")
```

```python
## Export as a tabular format
bangladesh_df.to_tsv("../../datas/bangladeshi_developers_survey.tsv", sep="\t")
```

```python
## Export Excel file 
## Step 1: Need to install: pip install xlwt openpyxl xlrd
bangladesh_df.to_excel("data/bangladeshi_developers_survey.xlsx")
```

```python
## Import Data from Excel
excel_df = pd.read_excel("data/bangladeshi_developers_survey.xlsx")
excel_df.head()
```

```python
## Export Data on JSON Format

bangladesh_df.to_json("data/bangladeshi_developers_survey.json")  # Output as a Dictionary
# bangladesh_df.to_json("data/bangladeshi_developers_survey.json", orient="records", lines=True)  # Output as a List
```

```python
## Import data from JSON format
json_df = pd.read_json("data/bangladeshi_developers_survey.json")
json_df.head()
```

```python
## Export SQL 
# Step 1: Need to install: pip install SQLAlchemy psycopg2-binary

from sqlalchemy import create_engine
import psycopg2

engine = create_engine('postgresql://dbuser:dbpass@localhost:5432/sample_db')
bangladesh_df.to_sql('sample_table_for_bangladesh', engine, if_exists='replace')
```

```python
## Import SQL
sql_df = pd.read_sql('sample_table_for_bangladesh', engine, index_col='Respondent')
```
