```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
df_people = pd.DataFrame(people)
```

```python
df_people
```

```python
## Return the last name if the Last name is Doe 
df_people['last'] == "Doe" # Answer the TRUE, FALSE value
filt = df_people['last'] == "Doe"
df_people[filt]
```

```python
## Using the Loc function

filt = (df_people['last'] == "Doe") & (df_people['first'] == "John")
df_people.loc[filt, 'email']
```

```python
## Oposite of the output. Here ~ Show the Opposite Vaule

filt = (df_people['last'] == "Doe") & (df_people['first'] == "John")
df_people.loc[~filt, 'email']
```
