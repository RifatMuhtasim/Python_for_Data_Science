## Small Data set

```python
people = {
    "first": ["Rifat", "Masum", "Sayem"], 
    "last": ["Muhtasim", "Abdullah", "Ahmed"], 
    "email": ["rifatmuhtasim@tavlav.com", "abdullah_masum@gmail.com", "sayme_ahmed@email.com"]
}
```

```python
import pandas as pd
df_people = pd.DataFrame(people)
```

```python
df_people
```

```python
## Change all of the Columns Name

df_people.columns = ["first_name", "last_name", "email"]
```

```python
df_people
```

```python
## lowercase for the Column Name

df_people.columns = [x.lower() for x in df_people.columns]
df_people
```

```python
## Remove all of the space in Column Name
df_people.columns = df_people.columns.str.replace("_", "")
df_people
```

```python
df_people
```

```python
## Update the specific column name 

df_people.rename(columns={'first_name': 'first', 'last_name': 'last'}, inplace=True)
df_people
```

```python
## Change the specific Column Values

df_people.loc[2, 'last'] = "Rifat"
```

```python
df_people
```

```python
## Using Filter 

filt = (df_people['email'] == "rifatmuhtasim@tavlav.com")
df_people[filt]
```

```python
## Update Using the filter

filt = (df_people['email'] == "rifatmuhtasim@tavlav.com")
df_people.loc[filt, 'first'] = "Khandokar"
```

```python
df_people
```

```python
 ## Lowercase all of the Email Value - Multiple

df_people['email'] = df_people['email'].str.lower()
df_people
```

```python
## 4 Methods to Change Multiple Rows (apply, map, applymap, replace)
## Using Apply in Series Object

df_people['email'].apply(len)

def update_email(email):
    return email.upper()

df_people['email'].apply(update_email)

## Using Lambda function
df_people['email'] = df_people['email'].apply(lambda x: x.lower())
df_people
```

```python
## Show all of the Maximum value 

df_people.apply(lambda x: x.max())
```

```python
## Using applymap and make all of the value to lowercase

df_people.applymap(str.lower)
```

```python
## Using replace function

df_people['first'] = df_people['first'].replace({'Rifat': 'Khandokar'})
df_people
```
